```python
# Set up the data
# Link: https://github.com/llimllib/bostonmarathon/blob/master/results/2014/results.csv
# 2014 Boston Marathon Data

def setdata(filename):
    df = pd.read_csv(filename)
    
    # Strip and make column names lowercase so that they are easy to manage. 
    df.columns = df.columns.str.strip().str.lower()

    # Replace characters or remove spaces and make all words lowercase.
    df.columns = df.columns.str.replace(' ', '_').str.replace('/', '_').str.replace('(', '').str.replace(')', '')

    return df
```

```python
# define function that returns a dataframe with fraction and percent of missing 
# data for each column
def percent_missing (df):
    numer = df.isnull().sum() 
    denom = df.isnull().count()

    data= []
    i = 0
    for x,y in zip(numer,denom):
        data.append([numer.index[i],(Fraction(x,y)), x*100/y])
        i+=1
    
    df1 = pd.DataFrame(data,columns = ['Column_Name', 'Fraction of Missing Data','Percent'])
    return df1

# preview this function in action with our defined dataframe
percent_missing(bm2014_df)
```

## K-means

```python
X = model_df[time_columns].copy()
```

```python
X
```

```python
# Normalize the data.
X_norm = normalize(X)

# Reduce it to two components.
X_pca = PCA(2).fit_transform(X_norm)
```

```python
def var_kmean(kmean_num,X_pca=X_pca):
    # calculate predicted valus for Kmean after PCA reduces data to two components
    # how long does Kmeans after PCA take to run?
    %timeit KMeans(n_clusters=kmean_num, random_state=42).fit_predict(X_pca)
    
    model = KMeans(n_clusters=kmean_num, random_state=42).fit(X_norm)
    y_pred_km = KMeans(n_clusters=kmean_num, random_state=42).fit_predict(X_pca)
    labels = model.labels_
    silhouette = metrics.silhouette_score(X_norm, labels, metric='euclidean')
    print(f'K: {kmean_num}, silhouette: {silhouette}')
    
    # Plot the solution.
    plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels)
    plt.show()
    
    # how long does minibatch approach take to run?
    %timeit MiniBatchKMeans(init='random', n_clusters=kmean_num,batch_size=200).fit(X_pca)
    minibatchkmeans = MiniBatchKMeans(init='random', n_clusters=kmean_num,batch_size=200).fit(X_pca)
    predict_mini = minibatchkmeans.predict(X_pca)
    
    # Check the MiniBatch model against our earlier one.
    print('Comparing k-means and mini batch k-means solutions:')
    print(pd.crosstab(predict_mini, y_pred_km))
    
    return labels
```

```python
label1 = var_kmean(2,X_pca)
```

```python
for col in time_columns[:7]:
    plt.scatter(X[col], X['official'], c=list(label1))
    plt.xlabel(col)
    plt.ylabel('Official')
    plt.title(f'{col} Split vs Official Finish Time: K=2')
    plt.show()
```

It's not immediately clear how these two clusters were segregated. I don't see an obvious take away from the clusters except that perhaps the purple group are those who maintained a more consistent and lower pace compared to the yellow group.
